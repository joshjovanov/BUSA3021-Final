```python
# importing libaries
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme(style="whitegrid")

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
```

```python
feature_cols = np.array(['Minimum temperature (?C)', 'Maximum temperature (?C)', 'Rainfall (mm)', 'Speed of maximum wind gust (km/h)',
        '9am relative humidity (%)',
       '9am wind direction', '9am wind speed (km/h)', '9am MSL pressure (hPa)',
       '3pm relative humidity (%)', '3pm wind direction',
       '3pm wind speed (km/h)', '3pm MSL pressure (hPa)', 'Count'])
```

```python
plt.figure(figsize=(10,8))
dataplot = sns.heatmap(df[feature_cols].corr(), cmap="RdGy", annot=True)
  
# displaying heatmap
plt.show()
```

```python
X = df[feature_cols]
y = df['Count']

#  splitting the data into 70% train and 30% test datasets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=142)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
models = {
    LogisticRegression(): "Logistic Regression",
    RandomForestRegressor(): "Random Forest"
}

for model in models.keys():
    model.fit(X_train_scaled, y_train)
```
